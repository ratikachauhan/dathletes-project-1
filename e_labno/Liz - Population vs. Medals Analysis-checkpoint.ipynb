```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import os
import numpy as np
```

```python
#import and read athlete CSV files
pwd = os.path.abspath('.')
athlete_path = os.path.join(pwd, 'athlete_events (1).csv')

athlete = pd.read_csv(athlete_path, sep=',')
athlete.head()
```

```python
#check Dtype
athlete.info()
```

```python
#check for Null values
#https://stackoverflow.com/questions/26266362/how-to-count-the-nan-values-in-a-column-in-pandas-dataframe/26266451#26266451
print(athlete.isnull().sum())
```

```python
#drop Null values 
#https://stackoverflow.com/questions/13413590/how-to-drop-rows-of-pandas-dataframe-whose-value-in-a-certain-column-is-nan
medal_winners = athlete.dropna(subset = ['Medal'])
medal_winners
```

```python
#sort column by years / view dTypes and size of dataset
medal_winners.sort_values(by = 'Year', inplace =True)
medal_winners.info()
```

```python
#view 
medal_winners.head()
```

```python
#delete rows base on column values for data set so only olympic years required for analysis
#Source in class examples / https://stackoverflow.com/questions/15315452/selecting-with-complex-criteria-from-pandas-dataframe

medal_winners_final = medal_winners[(medal_winners['Year'] >= 1952) & (medal_winners['Year'] <= 2010)] 
medal_winners_final
```

```python
#drop unneeded columns
#source: https://pandas.pydata.org/pandas-docs/version/0.21/generated/pandas.DataFrame.drop.html
new_df = medal_winners_final.drop(['Games', 'Season', 'City', 'Event', 'Name', 'Sex', 'Height', 'Weight', 'Age', 'ID', 'Team', 'Sport'], axis = 1)
new_df.head()
```

```python
#get medal counts
#source = https://towardsdatascience.com/pandas-tips-and-tricks-33bcc8a40bb9
medal_counts = pd.DataFrame(new_df.groupby(['NOC', 'Year'])['Medal'].value_counts()).rename(columns={'Medal':'Medal_Count'}).reset_index()
medal_counts
```

```python
#drop medal column
#source: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop.html

medal_only_counts = medal_counts.drop(['Medal'], axis = 1)
medal_only_counts.head()
```

```python
#cumulative sum of medals per country per total year
#Source: https://stackoverflow.com/questions/22650833/pandas-groupby-cumulative-sum/22651188#22651188

medal_only_counts = medal_only_counts.groupby(['NOC', 'Year'])['Medal_Count'].sum().reset_index()
print(medal_only_counts)
medal_only_counts['Final Medal Count'] = medal_only_counts.groupby(['NOC', 'Year'])['Medal_Count'].apply(lambda x: x.cumsum())


print(medal_only_counts)
```

```python
#create new dataframe with Russian Federation, USSR, and Soviet Bloc Countries, and outlier year 1992-Code EUN
#source: https://stackoverflow.com/questions/19960077/how-to-filter-pandas-dataframe-using-in-and-not-in-like-in-sql

noc_id_ussr_russia = ['RUS', 'MDA','KAZ', 'ARM', 'AZE', 'BLR', 'EST', 'GEO', 'KGZ', 'LAT', 'LTU', 'TJK', 'TKM', 'UKR', 'UZB', 'EUN', 'URS']
olympic_medal_by_country = medal_only_counts[medal_only_counts['NOC'].isin(noc_id_ussr_russia)]
olympic_medal_by_country
```

```python
#Replace NOC names with USSR for charting
olympic_medal_by_country['NOC'] = olympic_medal_by_country['NOC'].replace(['RUS', 'MDA','KAZ', 'ARM', 'AZE', 'BLR', 'EST', 'GEO', 'KGZ', 'LAT', 'LTU', 'TJK', 'TKM', 'UKR', 'UZB', 'EUN', 'URS'], 'USSR')
olympic_medal_by_country
```

```python
#group by year for charting
olympic_medal_by_country = olympic_medal_by_country.groupby(['NOC', 'Year'])['Medal_Count'].sum().reset_index()
olympic_medal_by_country
```

```python
#drop years 1992 and before

ussr_medal_winners_final = olympic_medal_by_country[(olympic_medal_by_country['Year'] < 1992)]
ussr_medal_winners_final
```

```python
#https://www.geeksforgeeks.org/matplotlib-pyplot-savefig-in-python/
#linegraphs all in class examples
plt.plot(ussr_medal_winners_final['Year'], ussr_medal_winners_final['Medal_Count'], label="Soviet (USSR)", c='red')
first_year = ussr_medal_winners_final['Year'].min()
last_year = ussr_medal_winners_final['Year'].max()
plt.xticks(np.arange(first_year, last_year + 1, 2), rotation=90)
plt.title("Soviet(USSR) Total Medal Wins 1952-2010")
plt.xlabel("Olympic Years")
plt.ylabel("Medals (Hundreds)")
plt.savefig("Soviet_Medals.png")
plt.legend()
plt.show()
```

```python
#create new dataframe with just Soviet Bloc countries, excluding Russia and USSR
#source:https://stackoverflow.com/questions/57602668/pandas-isin-on-column-entries-containing-lists

noc_id_ex_blox = ['MDA','KAZ', 'ARM', 'AZE', 'BLR', 'EST', 'GEO', 'KGZ', 'LAT', 'LTU', 'TJK', 'TKM', 'UKR', 'UZB']
olympic_medal_by_ex_blox = medal_only_counts[medal_only_counts['NOC'].isin(noc_id_ex_blox)]
olympic_medal_by_ex_blox
```

```python
#replace NOC column values for charting
olympic_medal_by_ex_blox['NOC'] = olympic_medal_by_ex_blox['NOC'].replace(['MDA','KAZ', 'ARM', 'AZE', 'BLR', 'EST', 'GEO', 'KGZ', 'LAT', 'LTU', 'TJK', 'TKM', 'UKR', 'UZB'], 'Post-Soviet Countries')
olympic_medal_by_ex_blox
```

```python
#groupby for charting
olympic_medal_by_ex_blox = olympic_medal_by_ex_blox.groupby(['NOC', 'Year'])['Medal_Count'].sum().reset_index()
olympic_medal_by_ex_blox
```

```python
plt.plot(olympic_medal_by_ex_blox['Year'], olympic_medal_by_ex_blox['Medal_Count'], label='Post-Soviet Countries', c='darkblue')
first_year = olympic_medal_by_ex_blox['Year'].min()
last_year = olympic_medal_by_ex_blox['Year'].max()
plt.xticks(np.arange(first_year, last_year + 1, 2), rotation=90)
plt.title("Post-Soviet Countries Total Medal Wins 1992-2010")
plt.xlabel("Olympic Years")
plt.ylabel("Medals (Hundreds)")
plt.legend()
plt.savefig("Ex-Soviet_Medals.png")
plt.show()
```

```python
#create a dataframe for only russia post 1991
#source:https://stackoverflow.com/questions/57602668/pandas-isin-on-column-entries-containing-lists
noc_id_only_russia = ['RUS', 'EUN']
olympic_medals_only_russia = medal_only_counts[medal_only_counts['NOC'].isin(noc_id_only_russia)]
olympic_medals_only_russia.head()
```

```python
#replace EUN with RUS for Chart
#source: https://datatofish.com/replace-values-pandas-dataframe/
olympic_medals_only_russia['NOC'] = olympic_medals_only_russia['NOC'].replace(['EUN'], 'RUS')
olympic_medals_only_russia
```

```python
plt.plot(olympic_medals_only_russia['Year'], olympic_medals_only_russia['Medal_Count'], label='Russian Federation', c='crimson')
first_year = olympic_medals_only_russia['Year'].min()
last_year = olympic_medals_only_russia['Year'].max()
plt.xticks(np.arange(first_year, last_year + 1, 2), rotation=90)
plt.title("Russian Federation Total Medal Wins 1992-2010")
plt.xlabel("Olympic Years")
plt.ylabel("Medals (Hundreds)")
plt.legend()
plt.savefig("RussianFederationMedalsPost1991.png")
plt.show()
```

```python
plt.plot(ussr_medal_winners_final['Year'], ussr_medal_winners_final['Medal_Count'], label="Soviet (USSR)", c='red')
plt.plot(olympic_medal_by_ex_blox['Year'], olympic_medal_by_ex_blox['Medal_Count'], label='Post-Soviet Countries', c='darkblue')
plt.plot(olympic_medals_only_russia['Year'], olympic_medals_only_russia['Medal_Count'], label= "Russian Federation", c='crimson')
first_year = ussr_medal_winners_final['Year'].min()
last_year = olympic_medals_only_russia['Year'].max()
plt.title("Soviet(USSR), Russian, and Post-Soviet Total Medal Wins 1952-2010")
plt.xlabel("Olympic Years")
plt.ylabel("Medals (Hundreds)")
plt.xticks(np.arange(first_year, last_year + 1, 2), rotation=90)
plt.savefig("AllMedals.png")
plt.legend(loc = 'upper left')

```

```python
#source #https://www.w3schools.com/python/gloss_python_for_range.asp
#source https://stackoverflow.com/questions/252703/what-is-the-difference-between-pythons-list-methods-append-and-extend
olympic_years = [str(year) for year in range(1952, 1993, 4)] + [str(year) for year in range(1994, 2011, 2)]
target_columns = ['Index', 'Region, subregion, country or area *', 'Country code']
target_columns.extend(olympic_years)
pop_clean_df = population[target_columns]
pop_clean_df.head()
```

```python
# subset by country
#source https://stackoverflow.com/questions/33392369/subset-selection-using-pandas-isin-syntax
pop_id_ussr = [643, 498, 398, 51, 31, 112, 233, 268, 417, 428, 440, 762, 795, 804, 860]

pop_ussr = pop_clean_df[pop_clean_df['Country code'].isin(pop_id_ussr)]

pop_ussr
```

```python
#eliminate space on population numbers to prepare for calculations
#source https://stackoverflow.com/questions/42462530/how-to-replace-the-white-space-in-a-string-in-a-pandas-dataframe
pop_ussr[olympic_years] = pop_ussr[olympic_years].replace(' ', '', regex=True)

pop_ussr[olympic_years] = pop_ussr[olympic_years].apply(pd.to_numeric)

pop_ussr
```

```python
#create dictionary of values for population of pre 1991 USSR
#source #https://www.w3schools.com/python/gloss_python_for_range.asp
pre_1991_years = [str(yr) for yr in range(1952, 1991, 4)]
ussr_pre_1991_pop_sizes = {year: pop_ussr[year].sum() for year in pre_1991_years}
ussr_pre_1991_pop_sizes
```

```python
#create new table of only Russian Federation population data to prep for analysis of post-1991 russian population data
post_1991_russia = pop_ussr.loc[pop_ussr['Country code'] == 643] 

post_1991_russia
```

```python
#create dictionary of sum for population of post 1991 Russia by Olympic Year
#source #https://www.w3schools.com/python/gloss_python_for_range.asp
post_1991_years = [str(yr) for yr in range(1992, 2011, 2)]
post_1991_pop_russia = {year: post_1991_russia[year].sum() for year in post_1991_years}
post_1991_pop_russia
```

```python
#create new population table with only Soviet Bloc Countries to prep for analysis of post 
#Source: In class examples
post_1991_bloc = pop_ussr.loc[pop_ussr['Country code'] != 643] 

post_1991_bloc
```

```python
#Create dictionary of sum of population
#Source #https://www.w3schools.com/python/gloss_python_for_range.asp
post_1991_years = [str(yr) for yr in range(1992, 2011, 2)]
post_1991_pop_bloc = {year: post_1991_bloc[year].sum() for year in post_1991_years}
post_1991_pop_bloc
```
